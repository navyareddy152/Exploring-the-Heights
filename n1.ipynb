```python
import pandas as pd
import numpy as np

exped = pd.read_csv('data/exped_tidy.csv', encoding='ISO-8859-1')
peaks = pd.read_csv('data/peaks_tidy.csv', encoding='ISO-8859-1')

print("Expedition Data:")
display(exped.info())
display(exped.head())

print("\nPeaks Data:")
display(peaks.info())
display(peaks.head())
```

```python
def missing_values_report(df):
    missing = df.isnull().sum()
    missing = missing[missing > 0].sort_values(ascending=False)
    missing_percent = (missing / len(df)) * 100

    report = pd.DataFrame({
        'Missing Values': missing,
        'Percent Missing': missing_percent.round(2)
    })
    
    return report

# Example usage:
exped_missing_report = missing_values_report(exped)
peaks_missing_report = missing_values_report(peaks)

# Display the report
print("Expedition Data Missing Values:\n", exped_missing_report)
print("\nPeaks Data Missing Values:\n", peaks_missing_report)
```

```python
#Expedition Data – Drop Unnecessary Columns
exped_cleaned = exped.drop(columns=[
    'ROUTE3', 'ROUTE4', 'ASCENT2', 'ASCENT3', 'ASCENT4',
    'ACCIDENTS', 'ACHIEVMENT', 'ASCENT1', 'OTHERSMTS', 'PRIMID'
])
```

```python
#Peaks Data – Drop Unnecessary Columns
peaks_cleaned = peaks.drop(columns=[ 
    'REFERMEMO', 'TREKYEAR', 'PSMTNOTE', 'PHOTOMEMO'
])
```

```python
#Fill Categorical Columns with Defaults-exp

exped_cleaned['ROUTE2'].fillna('Unknown', inplace=True)
exped_cleaned['TERMNOTE'].fillna('No Details Provided', inplace=True)
exped_cleaned['APPROACH'].fillna('Not Recorded', inplace=True)
exped_cleaned['TERMDATE'].fillna('Not Available', inplace=True)
exped_cleaned['BCDATE'].fillna('Not Available', inplace=True)
exped_cleaned['COUNTRIES'].fillna('Not Specified', inplace=True)
exped_cleaned['SMTDATE'].fillna('No Summit', inplace=True)
exped_cleaned['ROUTE1'].fillna('Unknown', inplace=True)
exped_cleaned['AGENCY'].fillna('Not Specified', inplace=True)
exped_cleaned['CAMPSITES'].fillna('Not Recorded', inplace=True)
```

```python
#Fill Numeric Columns-exp
exped_cleaned['ROUTEMEMO'].fillna(-1, inplace=True)
exped_cleaned['SMTTIME'].fillna(-1, inplace=True)
```

```python
#Fill Categorical Columns with Defaults-peak
peaks_cleaned['PKNAME2'].fillna('None', inplace=True)
peaks_cleaned['RESTRICT'].fillna('Not Specified', inplace=True)
peaks_cleaned['PSMTDATE'].fillna('Not Available', inplace=True)
peaks_cleaned['PEXPID'].fillna('Unknown', inplace=True)
peaks_cleaned['PCOUNTRY'].fillna('Not Specified', inplace=True)
peaks_cleaned['PSUMMITERS'].fillna('None', inplace=True)
peaks_cleaned['LOCATION'].fillna('Unknown', inplace=True)
```

```python
#Fill Numeric Columns-peak
peaks_cleaned['PYEAR'].fillna(-1, inplace=True)
peaks_cleaned['PEAKMEMO'].fillna(-1, inplace=True)
```

```python
# Final check for remaining missing values-exped
def missing_report(df):
    missing = df.isnull().sum()
    return missing[missing > 0].sort_values(ascending=False)

print("Missing values in cleaned expedition data:")
print(missing_report(exped_cleaned))

print("Missing values in cleaned peaks data:")
print(missing_report(peaks_cleaned))
```

```python
#Question 1: Highest Average Peak Height
import matplotlib.pyplot as plt
import seaborn as sns
# Group and sort by height
himal_avg = peaks_cleaned.groupby('HIMAL_FACTOR')['HEIGHTM'].mean().sort_values(ascending=False).reset_index()

# Set style
sns.set(style='whitegrid', font_scale=1.1)

# Set wider figure to prevent overlap
plt.figure(figsize=(16, 8))

# Use color palette
palette = sns.color_palette("coolwarm", len(himal_avg))

# Create barplot
bar = sns.barplot(data=himal_avg, x='HIMAL_FACTOR', y='HEIGHTM', palette=palette)

# Annotate each bar
for index, row in himal_avg.iterrows():
    bar.text(index, row['HEIGHTM'] + 100, f"{row['HEIGHTM']:.0f}m",
             color='black', ha='center', va='bottom', fontweight='bold')

# Improve label readability
plt.xticks(rotation=40, ha='right')  # Rotate and right-align
plt.title('🏔️ Average Peak Height by Mountain Range in the Himalayas', fontsize=16, weight='bold', pad=20)
plt.xlabel('Mountain Range')
plt.ylabel('Average Height (meters)')
plt.tight_layout()
plt.show()
```
