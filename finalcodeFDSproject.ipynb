```python
import pandas as pd

# Load the datasets with encoding fix
exped_df = pd.read_csv("data/exped_tidy.csv", encoding='ISO-8859-1')
peaks_df = pd.read_csv("data/peaks_tidy.csv", encoding='ISO-8859-1')

# Basic info
print("Expeditions Dataset:")
exped_df.info()
print("\nPeaks Dataset:")
peaks_df.info()
```

```python
print("Expeditions Data head:")
exped_df.head()
```

```python
print("Peaks Data head:")
peaks_df.head()
```

```python
# Function to summarize missing values
def missing_summary(df, name):
    print(f"\n--- Missing Values in {name} Dataset ---")
    missing = df.isnull().sum()
    percent = (missing / len(df)) * 100
    summary = pd.DataFrame({'Missing Count': missing, 'Missing %': percent})
    summary = summary[summary['Missing Count'] > 0]
    print(summary.sort_values(by='Missing %', ascending=False))

# Apply to both datasets
missing_summary(exped_df, "Expeditions")
missing_summary(peaks_df, "Peaks")
```

```python
# Expeditions
exped_df_cleaned = exped_df.drop(columns=[
    'ROUTE3', 'ROUTE4', 'ASCENT3', 'ASCENT4', 'ASCENT2',
    'ACCIDENTS', 'ACHIEVMENT', 'ASCENT1', 'ROUTE2',
    'PRIMID', 'OTHERSMTS', 'TERMNOTE',
    'APPROACH', 'TERMDATE', 'SMTTIME', 'BCDATE',
    'ROUTEMEMO', 'COUNTRIES'
])

# Peaks
peaks_df_cleaned = peaks_df.drop(columns=[
    'REFERMEMO', 'TREKYEAR', 'PHOTOMEMO', 'PSMTNOTE',
    'PKNAME2', 'RESTRICT'
])
```

```python
# Expeditions
exped_df_cleaned['SMTDATE'] = exped_df_cleaned['SMTDATE'].fillna("Unknown")
exped_df_cleaned['AGENCY'] = exped_df_cleaned['AGENCY'].fillna("Unknown")
exped_df_cleaned['CAMPSITES'] = exped_df_cleaned['CAMPSITES'].fillna("Unknown")
exped_df_cleaned['ROUTE1'] = exped_df_cleaned['ROUTE1'].fillna("Unknown")

# Logical columns
logical_cols = exped_df_cleaned.select_dtypes(include='bool').columns
exped_df_cleaned[logical_cols] = exped_df_cleaned[logical_cols].fillna(False)

# Peaks
peaks_df_cleaned['PEAKMEMO'] = peaks_df_cleaned['PEAKMEMO'].fillna("Unknown")
peaks_df_cleaned['LOCATION'] = peaks_df_cleaned['LOCATION'].fillna("Unknown")
```

```python
# Recheck for missing values in the cleaned datasets
print("\nRemaining missing values in Expeditions dataset:")
print(exped_df_cleaned.isnull().sum()[exped_df_cleaned.isnull().sum() > 0])

print("\nRemaining missing values in Peaks dataset:")
print(peaks_df_cleaned.isnull().sum()[peaks_df_cleaned.isnull().sum() > 0])
```

```python
# Fill categorical columns with 'Unknown'
peaks_df_cleaned['PYEAR'] = peaks_df_cleaned['PYEAR'].fillna("Unknown")
peaks_df_cleaned['PEXPID'] = peaks_df_cleaned['PEXPID'].fillna("Unknown")
peaks_df_cleaned['PSMTDATE'] = peaks_df_cleaned['PSMTDATE'].fillna("Unknown")
peaks_df_cleaned['PCOUNTRY'] = peaks_df_cleaned['PCOUNTRY'].fillna("Unknown")

# Fill numeric field with 0
peaks_df_cleaned['PSUMMITERS'] = peaks_df_cleaned['PSUMMITERS'].fillna(0)
```

```python
# Final null check
print("\n✅ Any missing values left?")
print(peaks_df_cleaned.isnull().sum().sum())
```

```python
# Merge on PEAKID
merged_df = pd.merge(exped_df_cleaned, peaks_df_cleaned, on='PEAKID', how='left')

# Quick check
print("Merged DataFrame shape:", merged_df.shape)
print("\nSample merged data:")
print(merged_df[['EXPID', 'PEAKID', 'PKNAME', 'HEIGHTM', 'HIMAL_FACTOR', 'PSTATUS_FACTOR', 'OPEN']].head())
```

```python
# See all columns in the merged dataframe
print("\nAll available columns in merged dataset:")
print(merged_df.columns.tolist())

# Check unique values for a few key categorical columns
print("\nUnique HIMAL ranges:", merged_df['HIMAL_FACTOR'].unique())
print("\nUnique climbing statuses:", merged_df['PSTATUS_FACTOR'].unique())
print("\nUnique seasons:", merged_df['SEASON_FACTOR'].unique())
print("\nOxygen usage (O2USED):", merged_df['O2USED'].value_counts())
print("\nSuccess on Route 1 (SUCCESS1):", merged_df['SUCCESS1'].value_counts())
```

```python
print("▶ Success Rate vs Oxygen Use")
print(merged_df.groupby('O2USED')['SUCCESS1'].value_counts(normalize=True).unstack())
```

```python
print("\n▶ Average Peak Height by Mountain Range")
print(merged_df.groupby('HIMAL_FACTOR')['HEIGHTM'].mean().sort_values(ascending=False))
```

```python
print("\n▶ Most Common Termination Reasons")
print(merged_df['TERMREASON_FACTOR'].value_counts())
```

```python
print("\n▶ Deaths vs No Hired Personnel")
print(merged_df.groupby('NOHIRED')[['MDEATHS', 'HDEATHS']].mean())
```

## 📊 Visualization #1: Supplemental Oxygen vs Summit Success
### ✅ Research Question
How does the use of supplemental oxygen affect summit success rates in Himalayan expeditions?

### 📈 Plot Type
Grouped Bar Chart

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Prepare data: success rate by O2 usage
o2_success = merged_df.groupby('O2USED')['SUCCESS1'].value_counts(normalize=True).unstack()

# Reset index for plotting
o2_success = o2_success.reset_index().melt(id_vars='O2USED', var_name='Success', value_name='Proportion')

# Plot
plt.figure(figsize=(8, 5))
sns.barplot(data=o2_success, x='O2USED', y='Proportion', hue='Success')
plt.title('Summit Success Rate by Supplemental Oxygen Use')
plt.xlabel('Supplemental Oxygen Used')
plt.ylabel('Proportion of Expeditions')
plt.xticks([0, 1], ['No', 'Yes'])
plt.legend(title='Summit Success')
plt.tight_layout()
plt.show()
```

## 📊 Visualization #2: Average Peak Height by Mountain Range
### ✅ Research Question
Which mountain ranges in the Nepal Himalaya have the highest average peak heights?

### 📈 Plot Type
Horizontal bar chart

```python
# Prepare data
height_by_range = merged_df.groupby('HIMAL_FACTOR')['HEIGHTM'].mean().sort_values(ascending=True)
height_df = height_by_range.reset_index()
height_df.columns = ['HIMAL_FACTOR', 'Average_Height']

# Plot with hue for color scale
plt.figure(figsize=(10, 7))
sns.barplot(
    data=height_df,
    x='Average_Height',
    y='HIMAL_FACTOR',
    hue='HIMAL_FACTOR',      # use hue for color gradient
    palette='Blues_d',
    dodge=False,             # avoids bar split by hue
    legend=False             # no need for redundant legend
)
plt.title('Average Peak Height by Mountain Range')
plt.xlabel('Average Height (meters)')
plt.ylabel('Mountain Range')
plt.tight_layout()
plt.show()
```

## 📊 Visualization #3: Most Common Termination Reasons
### ✅ Research Question
What are the most common reasons that Himalayan expeditions are terminated?

This gives insight into why expeditions fail — whether due to weather, physical conditions, technical issues, or other causes.

### 📈 Plot Type
Bar Chart (Descending Order)



```python
# Count termination reasons
term_counts = merged_df['TERMREASON_FACTOR'].value_counts().sort_values(ascending=True)

# Filter out "Success" and replot
term_counts_filtered = term_counts.drop(labels=[
    'Success (main peak)',
    'Success (subpeak, foresummit)'
])

# Plot only failure reasons
plt.figure(figsize=(10, 6))
sns.barplot(x=term_counts_filtered.values, y=term_counts_filtered.index, palette='flare')
plt.title('Non-Success Expedition Termination Reasons')
plt.xlabel('Number of Expeditions')
plt.ylabel('Termination Reason')
plt.tight_layout()
plt.show()
```

## 💀 Visualization #4: Deaths vs Use of Hired Personnel
### ✅ Research Question
Do expeditions without hired personnel experience fewer or more fatalities compared to those that use them?

This can reveal how reliance on hired support impacts risk.

### 📈 Plot Type
Side-by-Side Bar Plot for:

MDEATHS (member deaths)

HDEATHS (hired deaths)
Grouped by: NOHIRED (True/False)



```python

# Prepare data
death_counts = merged_df.groupby('NOHIRED')[['MDEATHS', 'HDEATHS']].sum().reset_index()
death_counts['NOHIRED'] = death_counts['NOHIRED'].map({True: 'No Hired Personnel', False: 'Hired Personnel'})

# Melt the data
melted = death_counts.melt(id_vars='NOHIRED', var_name='Death Type', value_name='Deaths')

# Custom color palette (optional)
palette = {'MDEATHS': 'salmon', 'HDEATHS': 'skyblue'}

# Plot
plt.figure(figsize=(8, 5))
ax = sns.barplot(data=melted, x='NOHIRED', y='Deaths', hue='Death Type', palette=palette)

# Fix legend labels
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, ['Member Deaths', 'Hired Personnel Deaths'], title='Death Type')

# Axis labels and title
plt.title('Total Deaths vs Hired Personnel Usage')
plt.xlabel('Hired Personnel Used')
plt.ylabel('Total Deaths')
plt.tight_layout()
plt.show()
```

### 📊 Visualization #6: Summit Success Rate by Season
## ✅ Research Question
How does seasonality affect the success of Himalayan climbing expeditions?

This helps climbers and planners understand when conditions are most favorable.

## 📈 Plot Type
Grouped Bar Chart – shows proportion of SUCCESS1 outcomes (True/False) for each SEASON_FACTOR.



```python
# Prepare success rates by season
season_success = merged_df.groupby('SEASON_FACTOR')['SUCCESS1'].value_counts(normalize=True).unstack()

# Convert to long format for plotting
season_success = season_success.reset_index().melt(id_vars='SEASON_FACTOR', 
                                                    var_name='Success', 
                                                    value_name='Proportion')

# Sort seasons logically (if needed)
season_order = ['Spring', 'Summer', 'Autumn', 'Winter']
season_success['SEASON_FACTOR'] = pd.Categorical(season_success['SEASON_FACTOR'], 
                                                  categories=season_order, 
                                                  ordered=True)

# Plot
plt.figure(figsize=(8, 5))
sns.barplot(data=season_success, x='SEASON_FACTOR', y='Proportion', hue='Success', palette='coolwarm')
plt.title('Summit Success Rate by Season')
plt.xlabel('Season')
plt.ylabel('Proportion of Expeditions')
plt.legend(title='Summit Success')
plt.tight_layout()
plt.show()
```

## 📊 Advanced Visualization #1: Success Rate by Year and Season
### ✅ Research Question
How does summit success rate vary across years and seasons in Himalayan expeditions (2020–2024)?

### 📈 Plot Type
Heatmap (Rows = SEASON_FACTOR, Columns = YEAR, Color = Success Rate)



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Filter to valid data (remove rows where season or success is missing)
filtered = merged_df.dropna(subset=['SEASON_FACTOR', 'YEAR', 'SUCCESS1'])

# Group and calculate success rate
heatmap_data = (
    filtered.groupby(['SEASON_FACTOR', 'YEAR'])['SUCCESS1']
    .mean()
    .reset_index()
    .pivot(index='SEASON_FACTOR', columns='YEAR', values='SUCCESS1')
)

# Sort seasons logically
season_order = ['Spring', 'Summer', 'Autumn', 'Winter']
heatmap_data = heatmap_data.reindex(season_order)

# Plot
plt.figure(figsize=(8, 5))
sns.heatmap(heatmap_data, annot=True, fmt=".2f", cmap="YlGnBu", linewidths=0.5)
plt.title('Summit Success Rate by Season and Year')
plt.xlabel('Year')
plt.ylabel('Season')
plt.tight_layout()
plt.show()
```

## 📊 Advanced Visualization #2: Route Success Rates by Mountain Range
### ✅ Research Question
Which climbing routes have the highest success rates across different Himalayan mountain ranges?

This helps identify region-specific challenges or efficiencies in route choices.

### 📈 Plot Type
Faceted Bar Chart (one plot per HIMAL_FACTOR, bars = ROUTE1, height = % success)

```python
# Filter rows with known routes and success info
route_data = merged_df.dropna(subset=['ROUTE1', 'SUCCESS1', 'HIMAL_FACTOR'])

# Group by region and route
route_success = (
    route_data.groupby(['HIMAL_FACTOR', 'ROUTE1'])['SUCCESS1']
    .mean()
    .reset_index()
)

# Optional: Filter to routes that occur frequently (to remove one-off routes)
route_counts = route_data['ROUTE1'].value_counts()
common_routes = route_counts[route_counts >= 5].index  # threshold: 5 or more uses
route_success = route_success[route_success['ROUTE1'].isin(common_routes)]

# Limit to top regions with the most data
top_regions = route_data['HIMAL_FACTOR'].value_counts().head(6).index
route_success = route_success[route_success['HIMAL_FACTOR'].isin(top_regions)]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set style
sns.set(style="whitegrid")

# Plot: one facet per region
g = sns.catplot(
    data=route_success,
    x='SUCCESS1',
    y='ROUTE1',
    col='HIMAL_FACTOR',
    kind='bar',
    col_wrap=3,
    height=4,
    sharex=False,
    palette='viridis'
)

g.set_titles("{col_name}")
g.set_axis_labels("Success Rate", "Climbing Route")
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle("Success Rates by Climbing Route and Mountain Range")
plt.tight_layout()
plt.show()
```

## 📈 Time Series #3: Monthly Summit Success Rate by Year
### ✅ Research Question
How does summit success vary by month and year?

This helps reveal if certain months are consistently better (or worse), and whether that pattern holds across different years.



```python
merged_df['SMTDATE'] = pd.to_datetime(merged_df['SMTDATE'], errors='coerce')


# Drop rows with missing summit date or success value
monthly_success_df = merged_df.dropna(subset=['SMTDATE', 'SUCCESS1']).copy()

# Extract month number and year
monthly_success_df['Month_Num'] = monthly_success_df['SMTDATE'].dt.month
monthly_success_df['Year'] = monthly_success_df['SMTDATE'].dt.year
```

```python
# Group and calculate success rate
monthly_success_rate = (
    monthly_success_df.groupby(['Year', 'Month_Num'])['SUCCESS1']
    .mean()
    .reset_index()
)
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=monthly_success_rate,
    x='Month_Num',
    y='SUCCESS1',
    hue='Year',
    marker='o',
    palette='tab10'
)
plt.title('Monthly Summit Success Rate by Year')
plt.xlabel('Month')
plt.ylabel('Success Rate')
plt.xticks(ticks=range(1, 13), labels=[
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'
])
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(title='Year')
plt.tight_layout()
plt.show()
```

## ⚠️ Time Series #4: Monthly Death Count Trend (2020–2024)
### ✅ Research Question
How do death counts vary over time — by month and year — for both members and hired personnel?

This gives a clear view of when expedition risks are highest across the calendar.

### 📈 Plot Type
Monthly trend line for:

MDEATHS = Member deaths

HDEATHS = Hired personnel deaths

```python
# Filter rows with summit dates
death_df = merged_df.dropna(subset=['SMTDATE']).copy()

# Extract year and month
death_df['Year'] = death_df['SMTDATE'].dt.year
death_df['Month_Num'] = death_df['SMTDATE'].dt.month
```

```python
# Group and sum deaths
monthly_deaths = (
    death_df.groupby(['Year', 'Month_Num'])[['MDEATHS', 'HDEATHS']]
    .sum()
    .reset_index()
)

# Optional: Add month name for display
monthly_deaths['Month_Name'] = monthly_deaths['Month_Num'].apply(
    lambda x: pd.to_datetime(str(x), format='%m').strftime('%b')
)
```

```python
import matplotlib.patches as mpatches
import matplotlib.lines as mlines

plt.figure(figsize=(12, 6))

# Plot member deaths (solid red tones)
sns.lineplot(
    data=monthly_deaths,
    x='Month_Num',
    y='MDEATHS',
    hue='Year',
    marker='o',
    palette='Reds'
)

# Plot hired personnel deaths (dashed blue tones)
sns.lineplot(
    data=monthly_deaths,
    x='Month_Num',
    y='HDEATHS',
    hue='Year',
    marker='s',
    palette='Blues',
    linestyle='--'
)

# Custom legend handles
member_legend = mlines.Line2D([], [], color='red', marker='o', label='Member Deaths')
hired_legend = mlines.Line2D([], [], color='blue', marker='s', linestyle='--', label='Hired Personnel Deaths')

plt.title('Monthly Death Counts by Year (Members vs Hired Personnel)')
plt.xlabel('Month')
plt.ylabel('Number of Deaths')
plt.xticks(
    ticks=range(1, 13),
    labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
            'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(handles=[member_legend, hired_legend], title='Death Type')
plt.tight_layout()
plt.show()
```
